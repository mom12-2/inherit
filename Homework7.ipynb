```python
class Employee:
    def __init__(self, name, base_salary):
        self.name = name
        self.base_salary = base_salary

    def calculate_pay(self):
        return self.base_salary


class Manager(Employee):
    def __init__(self, name, base_salary, bonus):
        super().__init__(name, base_salary)
        self.bonus = bonus

    def calculate_pay(self):
        return self.base_salary + self.bonus


manager = Manager("sreymom", 1500, 100)
print(manager.name, "is Manager.")


class Developer(Employee):
    def __init__(self, name, base_salary, programming_language):
        super().__init__(name, base_salary)
        self.programming_language = programming_language

    def write_code(self):
        print(f"{self.name} is writing code in {self.python} and get salary: ${self.salary}")


developer = Developer("sreyroth", 500, "Python")


print("Manager salary: $",manager.calculate_pay())

print(developer.name, "is Developer.")
developer.write_code()
```
